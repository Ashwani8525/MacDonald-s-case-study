# ****Analyzing of the Perception of McDonald's Food yummy among Customers****

```python
#Quantitative Analysis of McDonald's Customer Perception of Food yummy
df.yummy.value_counts()
```

```python
# plotting the yummy value in catplot
sns.catplot(x= 'yummy', kind = 'count', data =df, order = df.yummy.value_counts().index)
```

```python
#analysing through pie chart
df['yummy'].value_counts().plot(kind ='pie', autopct='%1.1f%%', shadow = True, startangle = 90)
plt.show()
```

*Almost 55% customers find McDonald's yummy*

```python
#Gender-Based Analysis of McDonald's Food yummy Perception 

plt.rcParams['figure.figsize'] = (15,8)
sns.countplot(x= 'yummy',hue ='Gender' , data =df)
plt.grid()#adding grid lines
plt.show()
```

*The count plot indicates that there is a difference in the frequency of responses regarding the yumminess of McDonald's food between females and males, with more females reporting that they find McDonald's food yummy compared to males.*

```python
#Age-Based Analysis of McDonald's Food yummy Perception
sns.countplot( x= 'Age', hue = 'yummy',data =df )
for p in plt.gca().patches:
    plt.gca().annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                ha='center', va='bottom', color= 'black', fontsize=10)
plt.title("Customer finding McDonald's yummy with Age ")
plt.grid()
plt.show()
```

```python
#plotting the barplot for different age group who find McDonald's yummy
bins = [20, 30, 40, 50, 60, 70, 80]
labels = ['20-29', '30-39', '40-49', '50-59', '60-69', '70-80']

df['Age Bin'] = pd.cut(df['Age'], bins=bins, labels=labels)

grouped = df.groupby(['Age Bin', 'yummy']).size().reset_index(name='Count')

sns.set(style='darkgrid')
sns.barplot(x='Age Bin', y='Count', hue='yummy', data=grouped)
#for getting the total count above the bar
for p in plt.gca().patches:
    plt.gca().annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                ha='center', va='bottom', color= 'black', fontsize=10)
```

```python
#Percentage Analysis of McDonald's yummy Perception among Different Age Groups
yummy_df=pd.crosstab(df['Age Bin'], df['yummy'])
yummy_df['Percentage yummy']= yummy_df.apply(lambda x : x[1]*100/(x[0]+x[1]),axis =1)
print(yummy_df)
```

*Among those aged 20-40, McDonald's is perceived as particularly yummy, with those aged group 20-30 finding it the most delicious. However, as age increases beyond 40, the proportion of people who find McDonald's yummy decreases.*

```python
#Percentage Analysis of McDonald's healthy Perception among Different Age Groups
healthy_df=pd.crosstab(df['Age Bin'], df['healthy'])
healthy_df["Percentage who find McDonald's healthy"]= healthy_df.apply(lambda x : x[1]*100/(x[0]+x[1]),axis =1)
print(healthy_df)
```
